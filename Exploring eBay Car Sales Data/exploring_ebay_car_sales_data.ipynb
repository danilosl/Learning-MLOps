Agora vamos fazer uma exploração dos dados básicos para determinar quais outras tarefas de limpeza precisam ser feitas.

```python
autos.describe(include='all')
```

Após a análise dos dados, notou-se que as colunas nr_pictures, offer_type e seller apresentam muitos valores correspondentes e nulos portanto, para preservar nossas analises futuras, elas devem ser removidas. Além disso, price e odometer são valores numéricos armazenados como texto.

```python
#Removendo as colunas com valores semelhantes e nulos.
autos = autos.drop(columns=['nr_pictures','offer_type','seller'])

#Remoção dos caracteres extras e conversão das colunas númericas armazenadas em texto.
autos['price'] = (autos['price'].str.replace('$','').str.replace(',', '').astype(float))
autos['odometer'] = (autos['odometer'].str.replace('km','').str.replace(',', '').astype(float))

#No caso da coluna odometer nao podemos perder a informacao que os valores sao medidos em km, vamos adicionar a informacao ao novo da coluna.
autos.rename(columns={"odometer":'odometer_km'},inplace=True)
```

Como um carro não pode ser registrado depois que a listagem foi vista, qualquer veículo com ano de registro maior que 2016 deve ser removido. Já determinar o primeiro ano válido é mais complicado, vamos tomar como partida o ano de 1900.  

```python
#Eliminando os outliers e apresentando a distribuição dos valores restantes.
autos = autos[autos.registration_year.between(1910,2016)]
autos.registration_year.value_counts(normalize=True)
```

Ao trabalharmos com dados sobre carros, é importante análisarmos variações entre diferentes marcas. Podemos usar a técnica de agregação entender a coluna  brand.

```python
autos['brand'].value_counts().head(20)
```

```python
#Criando uma serie com as 20 marcas mais populares da lista.
top_brand=autos.brand.value_counts().head(20).index
top_brand
```

```python
#Criando os dicionários para os preços médios e as quilometragens médias das top 20 marcas.
dict_mean_price={}
dict_mean_odometer={}
#Interando sobre eles.
for brand in top_brand:
    mean_price=autos[autos['brand']==brand]['price'].mean()
    mean_odometer=autos[autos['brand']==brand]['odometer_km'].mean()
    dict_mean_price[brand]=mean_price
    dict_mean_odometer[brand]=mean_odometer
```

```python
#Convertendo os dicionários em séries de objetos.
series_mean_price=pd.Series(dict_mean_price)
series_mean_odometer=pd.Series(dict_mean_odometer)
```

```python
#Criando um dataframe com a serie de objetos de preços médios.
df_relation_brand=pd.DataFrame(series_mean_price,columns=['mean_price'])
df_relation_brand
```

```python
#Atribuindo a serie de objetos de kilometragens médios, como uma nova coluna.
df_relation_brand['mean_odometer_km']=series_mean_odometer
df_relation_brand.sort_index()
```

Após a análise dos preços médios e quilometragens méidas das 20 marcas mais comuns,  não podemos perceber uma relação direta entre os dois valores.
