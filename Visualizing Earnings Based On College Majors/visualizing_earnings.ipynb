Neste projeto guiado, iremos explorar como usar as funcionalidades de plotagem, juntamente com as t√©cnicas forncidas pela biblioteca pandas em um conjunto de dados sobre os resultados do trabalho de alunos que se formaram na faculdade entre 2010 e 2012, gerados pela American Community Survey.

```python
# importando as bibliotecas necess√°rias
import pandas as pd
import matplotlib.pyplot as plt

# comando m√°gico para que os gr√°ficos sejam exibidos em linha
%matplotlib inline

# lendo o conjunto de dados
recent_grads = pd.read_csv('recent-grads.csv')

# retorna a primeira linha formatada como uma tabela
print(recent_grads.iloc[0])

# retorna os 5 primeiros elementos de cada coluna do dataframe
print(recent_grads.head())

# retorna os 5 ultimos elementos de cada coluna do dataframe
print(recent_grads.tail())

# retorna uma resumo geral sobre todas as colunas num√©ricas
recent_grads.describe()
```

```python
# descartando linhas com valores nulos
recent_grads = recent_grads.dropna()
```

### Gr√°ficos de Dispers√£o

```python
# Sample_size vs Median
recent_grads.plot(x = 'Sample_size', y = 'Median', kind = 'scatter')
```

```python
# Sample_size vs Unemployment_rate
recent_grads.plot(x = 'Sample_size', y = 'Unemployment_rate', kind = 'scatter')
```

```python
# Full_time vs Median
recent_grads.plot(x = 'Full_time', y = 'Median', kind = 'scatter')
```

```python
# ShareWomen vs Unemployment_rate
recent_grads.plot(x = 'ShareWomen', y = 'Unemployment_rate', kind = 'scatter')
```

```python
# Men vs Median
recent_grads.plot(x = 'Men', y = 'Median', kind = 'scatter')
```

```python
# Women vs Median
recent_grads.plot(x = 'Women', y = 'Median', kind = 'scatter')
```

### Histogramas

```python
# definindo as colunas necess√°rias
colunas = ["Sample_size", "Median", "Employed", "Full_time", 
           "ShareWomen", "Unemployment_rate", "Men", "Women"]

# gerando uma figura e definindo seu tamanho
fig = plt.figure(figsize = (5,20))

# realizando a intera√ß√£o entre o dataframe e nossa serie de colunas
for i in range(0,4):
    ax = fig.add_subplot(4, 1, i+1)
    ax = recent_grads[colunas[i]].plot(kind='hist', rot = 30)
```

### Gr√°fico de Matriz de Dispers√£o
Essa fun√ß√£o faz parte do mod√∫lo pandas.plotting e precisa ser importada separadamente.

```python
# importando a fun√ß√£o scatter matrix
from pandas.plotting import scatter_matrix

# criando uma gr√°fico de matriz de dispers√£o 2x2
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize = (8,8))
```

```python
# criando uma gr√°fico de matriz de dispers√£o 3x3
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize = (12,12))
```

### Gr√°fico de Barras

```python
# compara√ß√£o curso vs porcentagens de mulheres das primeiras 10 linhas
recent_grads.head(10).plot.bar(x = 'Major', y = 'ShareWomen', legend = False)
```

```python
# compara√ß√£o curso vs porcentagens de mulheres das √∫ltimas 10 linhas
recent_grads.tail(10).plot.bar(x = 'Major', y = 'ShareWomen', legend = False)
```

```python
# compara√ß√£o curso vs taxa de desemprego das primeiras 10 linhas
recent_grads.head(10).plot.bar(x = 'Major', y = 'Unemployment_rate', legend = False)
```

```python
# compara√ß√£o cursos vs taxa de desemprego das √∫ltimas 10 linhas
recent_grads.tail(10).plot.bar(x = 'Major', y = 'Unemployment_rate', legend = False)
```
